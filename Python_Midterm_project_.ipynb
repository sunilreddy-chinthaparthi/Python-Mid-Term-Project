```python
#to drop unimportant columns and rows
df.drop(['Unnamed: 0', 'Title', 'Review Text', 'Positive Feedback Count', 'Unnamed: 11', 'Unnamed: 12'],axis=1, inplace=True)
```

```python
#the amount of missing values per feature
sns.set(rc={'figure.figsize':(11,4)})
pd.isnull(df).sum().plot(kind='bar')
plt.ylabel('Number of missing values')
plt.title('Missing Values per Feature')
```

```python
# Dropping records having null values
df.dropna(inplace=True)
```

```python
# Checking if null values are removed
df.isnull().sum()
```

```python
# Checking if any duplicate records are present
duplicate=df[df.duplicated()]
duplicate
```

```python
# Removing duplicate records
df.drop_duplicates(inplace=True)
```

```python
# Again check if any duplicate records are left
duplicate = df[df.duplicated()]
duplicate
```

```python
df.rename(columns={'Recommended IND':'Recommended',
'Division Name':'Division', 'Department Name':'Department',
'Class Name ':' Product name'}, inplace=True)
```

```python
# Calculate summary statistics for numerical columns
numerical_cols = df.select_dtypes(include=np.number).columns
df[numerical_cols].describe()
```

```python
recommended = df[df['Recommended']==1]
not_recommended = df[df['Recommended']==0]
```

```python
#ax3 = plt.subplot2grid((2, 2), (1, 0), colspan=2)
plt.figure(figsize=(10,10))
ax3 = plt.xticks(rotation=45)
sns.countplot(recommended[' Product name'], color="cyan", label = "Recommended")
sns.countplot(not_recommended[' Product name'], color="blue",label = "Not Recommended")
plt.title("Recommended Items in each Prodect name ")
ax3 = plt.legend()
```

```python
# Create a scatter plot to visualize the relationship between age and product preferences
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y=' Product name', data=df)
plt.xlabel('Age of Reviewers')
plt.ylabel('Product Reviewed')
plt.title('Relationship between Age and Product Preferences')
plt.show()
```

```python
# Calculate the average rating for each product
product_ratings = df.groupby(' Product name')['Rating'].mean()

# Create a bar chart to visualize the average ratings
plt.figure(figsize=(10, 6))
sns.barplot(x=product_ratings.index, y=product_ratings.values)
plt.xlabel('Product')
plt.ylabel('Average Rating')
plt.title('Average Ratings for Each Product')
plt.xticks(rotation=45)
plt.show()

# Calculate the percentage of recommended products
recommended_percentage = df['Recommended'].value_counts(normalize=True) * 100

# Create a pie chart to visualize the percentage of recommended products
plt.figure(figsize=(6, 6))
plt.pie(recommended_percentage, labels=['Recommended', 'Not Recommended'], colors=['green', 'red'], autopct='%1.1f%%')
plt.title('Percentage of Recommended Products')
plt.show()
```

```python
# Calculate the correlation matrix
corr_matrix = df.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Features')
plt.show()
```
